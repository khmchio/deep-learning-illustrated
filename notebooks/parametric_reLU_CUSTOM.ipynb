```python
import matplotlib.pyplot as plt
import numpy as np

# Define ReLU and Leaky ReLU functions
def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha):
    return np.maximum(alpha * x, x)

# Define ranges for x-axis and alpha values
x = np.linspace(-2, 5, 100)
alpha_values = [0.01, 0.1, 0.5]

# Plot ReLU for reference
plt.plot(x, relu(x), label='ReLU')

# Display the plot
plt.grid(True)
plt.show()
```

```python

# Plot Leaky ReLU with different alpha values
for alpha in alpha_values:
    plt.plot(x, leaky_relu(x, alpha), label=f'Leaky ReLU (alpha={alpha})')

# Add labels and title
plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.title('ReLU vs. Leaky ReLU with Different Alpha Values')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
